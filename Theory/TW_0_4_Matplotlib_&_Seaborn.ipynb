#### Графики по умолчанию

*Разбор данного раздела:* https://youtu.be/Dk416j9CpK8?t=1274

<a href="http://matplotlib.org/">matplotlib</a> - библиотека для визуализации данных, в основном двухмерная графика.

```
import matplotlib.pyplot as plt
%matplotlib inline
```

Отрисовка графика "по умолчанию":

```
# создаем синтетические данные
import numpy as np 
x = np.linspace(0,15,100) # Возвращает равномерно распределенные числа в указанном интервале
print(x.shape)
print(x)
```

```
# рисуем график по умолчанию
plt.plot(x)
plt.show()
```

```
# график по умолчанию 2
y = np.cos(x) # задаем значения, соответсвующие функции косинус
plt.plot(y)
plt.show()
```

```
# график по умолчанию 3
y = np.cos(x) # задаем значения, соответсвующие функции косинус
plt.plot(x) 
plt.show()
```

Чем плох график выше?

* отсутствуют подписи осей
* отсутствует название графика

Давайте это поправим:

```
plt.plot(x, y) # указываем, какие значения выводим
plt.title('cos(x)') # название нашего графика
plt.xlabel('х') # название оси х
plt.ylabel('y') # название оси y
plt.show()
```

Добавим лейблы на график:

```
z = 0.5 * np.sin(x) # задаем значения, соответсвующие функции синус

# рисуем график
plt.plot(x, y, label='cos(x)') # указываем значения, по которым выводим график, указываем легенду в графе label
plt.plot(x, z, label='0.5 * sin(x)') # указываем значения, по которым выводим график, указываем легенду в графе label
plt.title('0.5 sin(x) и cos(x)') # название нашего графика
plt.xlabel('х') # ось х
plt.ylabel('y, z') # ось y,z
plt.legend() # вызываем объект легенда, чтобы на графике отображалась легенда
plt.show()
```

```
z = 0.5 * np.sin(x)

# рисуем график
plt.plot(x, y, label='cos(x)') # указываем значения, по которым выводим график, указываем легенду в графе label
plt.plot(x, z, label='0.5 * sin(x)') # указываем значения, по которым выводим график, указываем легенду в графе label
plt.title('0.5 sin(x) и cos(x)') # название нашего графика
plt.xlabel('х') # ось х
plt.ylabel('y, z') # ось y,z
plt.legend(loc='upper left') # вызываем объект легенда, чтобы на графике отображалась легенда, указываем параметр отображения этой легенды
plt.show()
```

```
z = 0.5 * np.sin(x)

# рисуем график
plt.plot(x, y, label='cos(x)')
plt.show()
plt.plot(x, z, label='0.5 * sin(x)')
plt.title('0.5 sin(x) и cos(x)')
plt.xlabel('х')
plt.ylabel('y, z')
plt.legend()
plt.show()
```

Все равно что-то не так... Что? 

Давайте поменяем размер!

```
# рисуем график
# заранее объявим объект для регулирования размер полотна для отображения графика
plt.figure(figsize=(12,4)) # параметры в скобках определяют размер графика 
# следующие три строки будут выводить три графика функции на одном полотне
plt.plot(x, y, label='cos(x)') 
plt.plot(x, z, label='0.5 * sin(x)')
plt.plot(x, 4*z, label='2 * sin(x)')
plt.title('Синусоиды')
plt.xlabel('х')
plt.ylabel('y, z, 4z')
plt.legend()
plt.show()
```

```
# рисуем график
plt.plot(x, y, label='cos(x)') # вывод графика до объявления размера полотна
plt.figure(figsize=(12,4)) # объявили размер полотна
# все остальные графики выводятся на нем
plt.plot(x, z, label='0.5 * sin(x)')
plt.plot(x, 4*z, label='2 * sin(x)')
plt.title('Синусоиды')
plt.xlabel('х')
plt.ylabel('y, z, 4z')
plt.legend()
plt.show()
```

Как поменять цвет?

```
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, y, label='cos(x)', c='green') # при вызове .plot в качестве параметра указываем цвет 'green' - зеленый
plt.plot(x, z, label='sin(x)', c='r') # при вызове .plot в качестве параметра указываем цвет 'red' - красный
plt.title('0.5 sin(x) и cos(x)') # название графика
plt.xlabel('х')
plt.ylabel('y, z')
plt.legend()
plt.show()
```

#### Варианты отрисовки

* прерывистая линия

```
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, z, 'b--') # для прерывистой голубой линии указываем параметр 'c--'
plt.title('0.5 sin(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
```

* звездочками

```
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, z, 'y*') # для вывода желтыми звездочками 'y*'
plt.title('0.5 sin(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
```

```
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, z, 'gs') # для вывода зелеными квадратами - 'gs'
plt.title('0.5 sin(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
```

```
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, z, 'c^') # для вывода голубыми треугольниками 'c^'
plt.title('0.5 sin(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
```

```
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, z, 'co') # для вывода голубыми кружками 'co'
plt.title('0.5 sin(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
```

Параметр linestyle:

```
styles = ['-', '--', '-.', ':', 'steps']

for ls in styles:
    # рисуем график
    plt.figure(figsize=(15,6))
    plt.plot(x, z, c='r', ls=ls) # т.к. в предыдущих примерах мы укзаывали цвет первой буквой и потом стиль линии
                                 # то в этом случае, когда хотим менять только стиль линии, а цвет оставить тем же
                                 # будем подставлять разные значения из списка стилей линий в параметр ls = 
    plt.title('0.5 sin(x)')
    plt.xlabel('х')
    plt.ylabel('y')
    plt.show()
```

Параметр marker:

```
styles = ['+', ',', '.', '1', '2', '3', '4' ]

for marker in styles:
    # рисуем график
    plt.figure(figsize=(15,6))
    plt.plot(x, z, c='r', marker=marker, ls=':', mec='b') # для параметра mec = указываем цвет маркера, а в marker, соответственно, тип маркера
    plt.title('0.5 sin(x)')
    plt.xlabel('х')
    plt.ylabel('y')
    plt.show()
```

Контроль толщины линии графика осуществляется за счет параметра linewidth:

```
# рисуем график
plt.figure(figsize=(15,6))
plt.plot(x, z, c='cornflowerblue', linewidth=2) # 'cornflowerblue'- все еще параметр цвета, linewidth = 2 - толщина 2 пикселя
plt.title('0.5 sin(x)')
plt.xlabel('х')
plt.ylabel('y')
plt.show()
```

## Несколько графиков в одних осях

*Разбор данного раздела:* https://youtu.be/Dk416j9CpK8?t=2984

```
plt.figure(1, figsize=(10,5)) # задаем размер полотна

plt.subplot(211) # указываем, что хотим создать дополнительный подграфик: 2 - количество строк, 1 - кол-во столбцов, 1 - какой по счету этот график
plt.title('Два графика с общей осью х')
plt.plot(x, y, 'b-.', label='cos(x)')
plt.legend()

plt.subplot(212) # указываем, что хотим создать дополнительный подграфик: 2 - количество строк, 1 - кол-во столбцов, 2 - какой по счету этот график
plt.plot(x, z, 'r--', label='0.5 sin(x)')
plt.xlabel('общая ось х')
plt.legend()

plt.show()
```

```
plt.figure(1, figsize=(10,5)) # задаем размер полотна

# Альтернативная равносильная запись создания подграфиков 
plt.subplot(2, 1, 1) # указываем, что хотим создать дополнительный подграфик: 2 - количество строк, 1 - кол-во столбцов, 1 - какой по счету этот график
plt.title('Два графика с общей осью х')
plt.plot(x, y, 'b-.', label='cos(x)')
plt.legend()

plt.subplot(2, 1, 2) # указываем, что хотим создать дополнительный подграфик: 2 - количество строк, 1 - кол-во столбцов, 1 - какой по счету этот график
plt.plot(x, z, 'r--', label='0.5 sin(x)')
plt.xlabel('общая ось х')
plt.legend()

plt.show()
```

```
plt.figure(1, figsize=(10,5))

plt.subplot(221)
plt.title('Два графика с общей осью х')
plt.plot(x, y, 'b-.', label='cos(x)')
plt.legend()

plt.subplot(222)
plt.plot(x, z, 'g--', label='0.5 sin(x)')
plt.xlabel('общая ось х')
plt.legend()

plt.subplot(224)
plt.plot(x, z, 'r--', label='0.5 sin(x)')
plt.xlabel('общая ось х')
plt.legend()

plt.show()
```

```
# Создаем игрушечные данные:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Задаем пространство для подграфиков
f, axes = plt.subplots(2, 2, sharex=False, sharey=False, figsize=(12,8))

# График с координатами положения (0,0) по отношению к оси 
axes[0,0].plot(x,y,label='sin(x^2)', c='r')
axes[0,0].legend() 

# График с координатами положения (0,1) по отношению к оси 
axes[0,1].plot(x, np.abs(y), label='np.abs(y)')
axes[0,1].legend()

# График с координатами положения (1,0) по отношению к оси 
axes[1,0].plot(x,y**3)

# График с координатами положения (1,1) по отношению к оси 
axes[1,1].plot(x,np.arccos(y))
plt.show()
```
